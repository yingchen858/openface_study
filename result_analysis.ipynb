```python
import pickle
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
with open('similarity_result.pkl', 'rb') as f:
   similarity_result = pickle.load(f)

with open('truth.pkl', 'rb') as f:
   truth = pickle.load(f)
```

```python
len(similarity_result), len(truth)
```

```python
zip(similarity_result,truth)
```

```python
#precision, recall, thresholds = precision_recall_curve(y, pred)
fpr1, tpr1, thresholds1 = metrics.roc_curve(truth, similarity_result, pos_label=1)

lw =2 
plt.clf()
##plt.plot(fpr, tpr, lw=lw, color='navy',
##         label='geohash3_w2v_10d')
##plt.plot(fpr2, tpr2, lw=lw, color='green',
##         label='word2vec_20d')
plt.plot(fpr1, tpr1,'b-',lw=lw,label='similarity_same_person')
##plt.plot(fpr4, tpr4, lw=lw, color='yellow',
##         label='word2vec_3d')


plt.xlim(0,1.0)
plt.ylim(0,1.0)
plt.minorticks_on()
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='g', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc=(1.05, 0.65))
plt.show()

```

```python
density_input_same=[]
density_input_different=[]
for x in zip(similarity_result, truth):
    if x[1]==1:
        density_input_same.append(x[0])
    else:
        density_input_different.append(x[0])
```
